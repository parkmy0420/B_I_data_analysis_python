```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# groupby(): 데이터를 그룹별로 나누고 요약 및 집계할 수 있다. 
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
# clarity(다이아몬드의 투명도) 컬럼 기준으로 그룹화한 후 price 컬럼의 평균값 파악
grouped = diamonds['price'].groupby(diamonds['clarity'])
grouped
```

```python
# 투명도별 다이아몬드의 평균 가격(mean: 평균)
grouped.mean()
```

```python
# 상위 인덱스 color, 하위 인덱스 clarity 2개 레벨로 구성된 인덱스를 가진 평균 가격 데이터
grouped_2 = diamonds['price'].groupby([diamonds['color'], diamonds['clarity']]).mean()
grouped_2
```

```python
grouped_2.unstack()
```

```python
# cut 컬럼 인덱스로 전체 데이터의 평균을 구하는 그룹 연산
# 결과 - 범주형 컬럼인 color, clarity를 제외한 모든 컬럼의 평균 가격 확인 가능(숫자가 아닌 값은 자동으로 제외됨)
diamonds.groupby(['cut']).mean()
```

```python
# size()는 결과를 시리즈로 반환, unstack과 결합하면 데이터프레임으로 반환
diamonds.groupby(['cut', 'clarity']).size()
```

```python
# unstack : 하위 인덱스가 컬럼으로 변환덴 데이터 프레임을 반환
diamonds.groupby(['cut', 'clarity']).size().unstack()
```

```python
# 일부 컬럼의 그룹 연산 확인
# cut컬럼 기준으로 price의 평균을 구하는 연산
diamonds.groupby('cut')['price'].mean()
```

```python
# color컬럼 기준으로 carat의 평균을 구하는 연산
diamonds.groupby('color')['carat'].mean()
```

```python
diamonds.groupby(['cut', 'clarity'])['price'].mean().unstack()
```

```python
# 컬럼에 다양한 함수를 적용하는 방법
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_percent'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
# 요일과 흡연 여부 기준에 따라 데이터 그루핑
tips_byday =tips.groupby(['day', 'smoker'])
```

```python
tips_byday_percent = tips_byday['tip_percent']
tips_byday_percent.agg('mean')
```

```python
tips_byday_percent.agg(['mean', 'std', 'sum'])
```

```python
# 함수명 대신 원하는 이름으로 출력 (원하는 이름, 함수명)
tips_byday_percent.agg([('팁_평균', 'mean'), ('팁_표준편차', 'std')])
```

```python
function_list = ['count', 'mean', 'max']
tips_byday['tip_percent', 'total_bill'].agg(function_list)
```

```python
function_tuple = [('팁_평균', 'mean'), ('팁_표준편차', 'std')]
tips_byday['tip_percent', 'total_bill'].agg(function_tuple)
```

```python
# 컬럼마다 다른 함수를 적용하고 싶을때 딕셔너리 사용
tips_byday.agg({'tip': 'max', 'size': 'sum'})
```

```python
# 집계하고 싶은 함수가 여러개인 경우 리스트로
tips_byday.agg({'tip_percent': ['min', 'max', 'mean', 'std'], 'size': 'sum'})
```

```python
# 다중형 인덱스 구조의 데이터프레임
tips.groupby(['day', 'smoker']).mean()
```

```python
# 일반적인 형태의 데이터 프레임(as_index=False)
tips.groupby(['day', 'smoker'], as_index=False).mean()
```

```python
# 위와 똑같은 결과값이 나오는 방식이지만 불필요한 계산을 거치는 방법
tips.groupby(['day', 'smoker']).mean().reset_index()
```

```python
# groupby_apply 작업(고급 데이터 변환 및 집계작업, 피벗 테이블 관련 작업)
penguins = sns.load_dataset('penguins')
penguins = penguins.dropna()
penguins.head()
```

```python
def top(data, n=5, column='body_mass_g'):
    return data.sort_values(by=column)[-n:]
```

```python
top(penguins, n=8)
```

```python
# sex 기준으로 그루핑 한 후 각각 body_mass_g의 상위 5개 데이터 반환
penguins.groupby('sex').apply(top)
```

```python
# penguins.groupby(['sex', 'species']).apply(top, n=1, column='body_mass_g')
penguins.groupby(['sex', 'species']).apply(top, 1)
```

```python
penguins.groupby(['sex', 'species']).apply(top, 1, 'bill_length_mm')
```

```python
# groupby, describe의 함수를 결합하면 좀 더 상세한 분포 확인 가능
# sex를 기준으로 body_mass_g값에 관한 다양한 통계량
penguins.groupby('sex')['body_mass_g'].describe()
```

```python
# 각 컬럼을 행으로 변환
penguins.groupby('sex')['body_mass_g'].describe().unstack('sex')
```

```python
# groupby, describe함수로 생성한 통계 결과를 기준으로 apply에 lambda를 활용하여 백분위수까지 반환
penguins.groupby('sex')['body_mass_g', 'flipper_length_mm'].apply(lambda x: x.describe())
```

```python
'''데이터를 구간화하는 cut, qcut과 groupby를 결합하면 특정 그룹 단위에 대한 버킷 분석을 쉽게 할 수 있음
   버킷 분석: 데이터를 버킷처럼 그룹별로 나누어 분석'''

# b_m_g변수를 5개 구간으로 나눈 값을 b_m_g_b이라는 새로운 컬럼 생성
penguins['body_mass_g_bin'] = pd.cut(penguins['body_mass_g'], bins=5)
penguins.head()
```

```python
penguins.body_mass_g_bin.value_counts()
```

```python
penguins['flipper_length_mm_bin'] = penguins.groupby('body_mass_g_bin').flipper_length_mm.apply(pd.cut, bins=2)
penguins.head()
```

```python
# 위와 같은 결과값 apply, transform 사용 차이
penguins['flipper_length_mm_bin'] = penguins.groupby('body_mass_g_bin')['flipper_length_mm'].transform(lambda x:pd.cut(x, bins=2))
penguins.head()
```

```python
penguins.flipper_length_mm_bin.value_counts()
```

```python
penguins['bill_length_mm_bin_by_island'] = penguins.groupby('island').bill_length_mm.apply(pd.cut, bins=2)
penguins[['island', 'bill_length_mm', 'bill_length_mm_bin_by_island']].head(10)
```
